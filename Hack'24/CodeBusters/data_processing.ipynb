```python
def process(str):
    str=str.lower().strip()
    if str[-1]=='.':
        str=str[:-1]
    return str
for key, value in data.items():
    data[key]=[process(val) for val in value]
```

```python
value_mapper={"Police":0, "Anti-Bribery":1, "Consumer Affairs":2, "Electricity":3, "Fire":4, "General":5, "MCD":6}
```

```python
#pre-processing text
words=[]
queries=0
for value in data.values():
    for val in value:
        queries+=1
        words+=val.split()

words=list(set(words))
word_mapper={}

for ind, word in enumerate(words):
    word_mapper[word]=ind

f=len(words)
features=np.zeros((queries, f+1))
outputs=np.zeros((queries,1))

i=0
for key, value in data.items():
    for val in value:
        outputs[i]=value_mapper[key]
        for word in val.split():
            features[i][word_mapper[word]]+=1
        i+=1

print(outputs)
```

```python
output_path="./queries.txt"

with open(output_path, 'w') as op:
    op.write(json.dumps({"outputs":list(outputs.flat), "features":[list(feature) for feature in features], "value-mapper":value_mapper, "word-mapper":word_mapper}))
```
